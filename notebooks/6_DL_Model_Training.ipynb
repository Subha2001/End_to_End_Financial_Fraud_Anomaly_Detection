# **Train data with different DL Algorithm**

```python
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

## **Artificial Neural Network**

```python
# Build ANN Model (Enhanced for Financial Fraud Detection)
def build_ann_model(input_dim):
    model = Sequential()
    model.add(Dense(128, input_dim=input_dim, activation='relu'))  # Increased neurons
    model.add(BatchNormalization()) # Batch normalization for stability
    model.add(Dropout(0.5))  # Dropout for regularization
    model.add(Dense(64, activation='relu')) # Increased neurons
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu')) # Added layer
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))  # Output layer

    model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]) #Lower learning rate, AUC
    return model

ann_model = build_ann_model(X_train_resampled[:3500000].shape[1])
```

```python
#Train model
ann_train = ann_model.fit(X_train_resampled[:3500000], y_train_resampled[:3500000], epochs=10, batch_size=32, validation_data=(X_val_transformed, y_val), class_weight={0: 1, 1: 8}) #increased epochs, batch_size, and class weights.
```

```python
#Evaluate the Model (Enhanced Evaluation)
y_pred_proba = ann_model.predict(X_val_transformed)
y_pred = (y_pred_proba > 0.5).astype(int)

loss, accuracy, auc = ann_model.evaluate(X_val_transformed, y_val)
print(f'Loss: {loss}, Accuracy: {accuracy}, AUC: {auc}')
print(classification_report(y_val, y_pred))
print(f"ROC AUC Score: {roc_auc_score(y_val, y_pred_proba)}")
```

```python
# Evaluate the performance on the validation data
print("Classification Report:")
print(classification_report(y_val, y_pred))

accuracy = accuracy_score(y_val, y_pred)
mcc = matthews_corrcoef(y_val, y_pred)

print(f"Validation Accuracy: {accuracy:.4f}")
print(f"Validation Matthews Correlation Coefficient: {mcc:.4f}")
```

## **AutoEncoder**

```python
# Build AutoEncoder Model (Enhanced for Anomaly Detection)
def build_autoencoder(input_dim, encoding_dim):
    input_layer = Input(shape=(input_dim,))
    encoder = Dense(128, activation='relu')(input_layer)
    encoder = BatchNormalization()(encoder)
    encoder = Dropout(0.2)(encoder)
    encoder = Dense(64, activation='relu')(encoder)
    encoder = BatchNormalization()(encoder)
    encoder = Dropout(0.2)(encoder)
    encoder = Dense(encoding_dim, activation='relu')(encoder) # bottleneck
    decoder = Dense(64, activation='relu')(encoder)
    decoder = BatchNormalization()(decoder)
    decoder = Dropout(0.2)(decoder)
    decoder = Dense(128, activation='relu')(decoder)
    decoder = BatchNormalization()(decoder)
    decoder = Dropout(0.2)(decoder)
    decoder = Dense(input_dim, activation='sigmoid')(decoder)

    autoencoder = Model(inputs=input_layer, outputs=decoder)
    autoencoder.compile(optimizer=Adam(learning_rate=0.0001), loss='mean_squared_error')
    return autoencoder

encoding_dim = 8 # increased encoding dimension.
autoencoder = build_autoencoder(X_train_resampled[:3500000].shape[1], encoding_dim)
```

```python
#Train model
autoencoder_train = autoencoder.fit(X_train_resampled[:3500000], y_train_resampled[:3500000], epochs=10, batch_size=32, validation_data=(X_val_transformed, y_val), class_weight={0: 1, 1: 8}) #Increased epochs and batch size.
```

```python
from sklearn.metrics import classification_report, accuracy_score, matthews_corrcoef
```

```python
#Evaluate the Model (Enhanced Anomaly Detection)
reconstructions = autoencoder.predict(X_val_transformed)
mse = np.mean(np.power(X_val_transformed - reconstructions, 2), axis=1)

threshold = np.percentile(mse, 99) # Increased threshold for stricter anomaly detection.
y_pred_autoencoder = (mse > threshold).astype(int)

mcc = matthews_corrcoef(y_val, y_pred_autoencoder)

print("Autoencoder Anomaly Detection:")
print(classification_report(y_val, y_pred_autoencoder))
print(f"Accuracy: {accuracy_score(y_val, y_pred_autoencoder)}")
print(f"Validation Matthews Correlation Coefficient: {mcc:.4f}")
print(f"ROC AUC Score: {roc_auc_score(y_val, mse)}") # MSE based AUC
```

# **Hyperparameter Tuning of Deep Learning Model [ANN]**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, matthews_corrcoef, accuracy_score
import random

# ------------------------------------------------------------
# 1. Use your existing train and validation data
# ------------------------------------------------------------
# In this example, we use the first 1,000,000 samples to define the feature space.
# (Feel free to adjust according to your memory and needs)
num_features = X_train_resampled[:100000].shape[1]

# For training, use the full training data you intend:
# (Here you might replace with your actual slicing; for demonstration using the full arrays)
X_train_final = X_train_resampled[:100000]
y_train_final = y_train_resampled[:100000]

# For validation, use the provided transformed data
X_val_final = X_val_transformed[:10000]
y_val_final = y_val[:10000]

# ------------------------------------------------------------
# 2. Build the model construction function
# ------------------------------------------------------------
def build_model_from_params(params):
    """
    Build a Keras sequential model based on the provided hyperparameters.

    params: A dictionary with keys:
      - hidden_layers: Number of hidden layers.
      - neurons: Number of neurons per hidden layer.
      - learning_rate: Learning rate for the optimizer.
      - dropout_rate: Dropout rate value.
      - activation: Activation function to use in hidden layers.
      - use_batchnorm: Boolean indicating whether to use BatchNormalization.
      - use_dropout: Boolean indicating whether to add a Dropout layer.
    """
    model = keras.Sequential()

    # Input layer
    model.add(keras.Input(shape=(num_features,)))

    # Hidden layers
    for i in range(params['hidden_layers']):
        model.add(layers.Dense(params['neurons'], activation=params['activation']))
        if params['use_batchnorm']:
            model.add(layers.BatchNormalization())
        if params['use_dropout'] and params['dropout_rate'] > 0.0:
            model.add(layers.Dropout(params['dropout_rate']))

    # Output layer for binary classification
    model.add(layers.Dense(1, activation='sigmoid'))

    # Compile the model
    optimizer = keras.optimizers.Adam(learning_rate=params['learning_rate'])
    model.compile(optimizer=optimizer,
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model

# ------------------------------------------------------------
# 3. Define the hyperparameter grid
# ------------------------------------------------------------
param_distributions = {
    'hidden_layers': [3, 4],
    'neurons': [64, 128, 256],
    'learning_rate': [0.001, 0.01, 0.1],
    'dropout_rate': [0.0, 0.2, 0.5],
    'activation': ['relu', 'tanh'],
    'use_batchnorm': [True],
    'use_dropout': [True, False],
}
```

```python
# ------------------------------------------------------------
# 4. Run 12 random combinations of hyperparameters
# ------------------------------------------------------------
num_trials = 8
results = []

print("Starting random hyperparameter search with 12 trials...\n")

for trial in range(num_trials):
    # Sample a random hyperparameter combination from the grid.
    sampled_params = {
        key: random.choice(val) for key, val in param_distributions.items()
    }

    print(f"Trial {trial + 1}/{num_trials} with parameters: {sampled_params}")

    # Build the model for this combination.
    model = build_model_from_params(sampled_params)

    # Train the model with 5 epochs and a batch size of 16.
    model.fit(X_train_final, y_train_final,
              epochs=5,
              batch_size=16,
              validation_data=(X_val_final, y_val_final),
              verbose=0,
              class_weight={0: 1, 1: 8})  # Set verbose=0 to suppress training output per trial

    # Evaluate the model on the validation data.
    y_pred_prob = model.predict(X_val_final)
    y_pred = (y_pred_prob > 0.5).astype(int)

    # Calculate performance metrics.
    acc = accuracy_score(y_val_final, y_pred)
    mcc = matthews_corrcoef(y_val_final, y_pred)
    results.append({
        'params': sampled_params,
        'accuracy': acc,
        'mcc': mcc
    })
    print(f"Trial {trial + 1}: Accuracy = {acc:.4f}, MCC = {mcc:.4f}\n")
```

```python
# ------------------------------------------------------------
# 5. Determine the best hyperparameter combination based on MCC
# ------------------------------------------------------------
best_model = max(results, key=lambda x: x['mcc'])
print("Best hyperparameter combination based on MCC:")
print(best_model['params'])
print(f"Best MCC: {best_model['mcc']:.4f}, with Accuracy: {best_model['accuracy']:.4f}")

# Optionally, print all results
print("\nAll trial results:")
for idx, result in enumerate(results, 1):
    print(f"Trial {idx}: Params: {result['params']}, Accuracy: {result['accuracy']:.4f}, MCC: {result['mcc']:.4f}")
```
